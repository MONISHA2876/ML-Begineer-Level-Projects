```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

**DATA COLLECTION**

```python
diabetes_data_raw = pd.read_csv('/content/drive/MyDrive/ML Models DataSet/diabetes.csv')
```

```python
diabetes_data_raw.head()
```

**ANALYSING DATA**

```python
diabetes_data_raw.shape
```

```python
diabetes_data_raw.value_counts('Outcome')
#Here the outcome 0 is almost double the outcome 1
#Here we have to handle this imbalanced data
#But this data set is really small removing some cells will make this dataset smaller
```

```python
#Check for missing values
diabetes_data_raw.isnull().sum()
```

```python
#Check discription o data
diabetes_data_raw.describe()
```

```python
diabetes_data_raw.groupby('Outcome').mean()
```

```python
#Seperating the labels from data
X = diabetes_data_raw.drop('Outcome',axis=1)
Y = diabetes_data_raw[['Outcome']]

print(X)
print(Y)
```

**DATA STANDARDIZATION**

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
print(X)
print(Y)
```

**Splitting training and testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

**TRAINING THE MODEL**

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

**EVALUATNG THE MODEL**

```python
accuracy_on_training_data = accuracy_score(classifier.predict(X_train),Y_train)
print('Accuracy on training data is ',accuracy_on_training_data)

accuracy_on_testing_data = accuracy_score(classifier.predict(X_test),Y_test)
print('Accuracy on testing data is ',accuracy_on_testing_data)
```

**MAKING THE PREDICTION SYSTEM**

```python
input = (6,144,72,27,228,33.9,0.255,40)

#Convert the input into numpy array
input_array = np.asarray(input)

#Reshaping the array
input_reshaped = input_array.reshape(1,-1)

#Standadising the input data
std_data = scaler.fit_transform(input_reshaped)

Prediction = classifier.predict(std_data)

if Prediction[0] == 0:
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
